## Veamos nuestros datos en un mapa

````python
import folium
````

## Primero: Tenemos que ubicar donde en qué **columnas** están los datos de :
* latitud
* longitud

````python
file_data.columns
````

En este caso las columnas con las **coordendas** son:
* 'Latitud en GD'
* 'Longitud en GD'

## Segundo vamos a generar un mapa **base** para luego agregar los puntos

````python
def generar_mapa():
    attr = (
        '&copy; <a href="https://www.openstreetmap.org/copyright">OpenStreetMap</a> '
        'contributors, &copy; <a href="https://cartodb.com/attributions">CartoDB</a>'
    )

    tiles = 'https://wms.ign.gob.ar/geoserver/gwc/service/tms/1.0.0/capabaseargenmap@EPSG%3A3857@png/{z}/{x}/{-y}.png'
    m = folium.Map(
        location=(-33.457606, -65.346857),
        control_scale=True,
        zoom_start=5,
        name='es',
        tiles=tiles,
        attr=attr
    )
    return m
````

Usamos la función **generar_mapa** para tener nuestro mapa base

````python
mapa = generar_mapa()
````

La variable **mapa** es nuestra mapa **base**

### Tercero agregamos a nuestro mapa:
En la celda de abajo pusimos el código para poder agregar los puntos, te contamos lo más importante para tener en cuenta.
* los puntos que se encuentran en las columnas de las **coordenadas**
```python
[row['Latitud en GD'], row['Longitud en GD']],
```
* queremos que muestre el nombre del lago cuando hacemos clic sobre cada icono, por eso pusimos el nombre de la columna donde se encuentra **ese dato**
```python
popup=row['Nombre']
```

````python
mapa = generar_mapa()
def agregar_marca_lago(row):
    folium.Marker(
        [row['Latitud en GD'], row['Longitud en GD']],
        popup=row['Nombre'],
        icon=folium.Icon()
        ).add_to(mapa)
file_data.apply(agregar_marca_lago, axis=1)
mapa
````

# 5. Armá tu mapa
Con el dataset elegido mostrá los datos en un mapa

## Primero encontrá los nombres de las columnas donde están las coordenadas

````python
file_data = pd.read_csv('ar-airports.csv')
file_data.columns
````

En base al resultado las coordenadas son 'latitude_deg' y 'longitude_deg'

## Segundo creá tu mapa **base**

````python
def generar_mapa():
    attr = (
        '&copy; <a href="https://www.openstreetmap.org/copyright">OpenStreetMap</a> '
        'contributors, &copy; <a href="https://cartodb.com/attributions">CartoDB</a>'
    )

    tiles = 'https://wms.ign.gob.ar/geoserver/gwc/service/tms/1.0.0/capabaseargenmap@EPSG%3A3857@png/{z}/{x}/{-y}.png'

    m = folium.Map(
        location=(-37.457606, -65.346857),
        control_scale=True,
        zoom_start=4.5,
        name='es',
        tiles=tiles,
        attr=attr
    )
    return m
````

## Tercero agregá los puntos a tu mapa

````python
mapa = generar_mapa()

def agregar_aeropuerto(row):

#Al pop up le agregué que muestre también los códigos IATA de los aeropuertos junto a sus nombres
  popup=row['name'] + ' ' + '(' + str(row['iata_code']) + ')'

  folium.Marker(
      [row['latitude_deg'], row['longitude_deg']],
      popup=popup,
#Y le cambié el color a los pines del mapa
      icon=folium.Icon(color='red')
      ).add_to(mapa)

file_data.apply(agregar_aeropuerto, axis=1)
mapa
````
